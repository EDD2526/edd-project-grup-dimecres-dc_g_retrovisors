```python
import matplotlib.pyplot as plt

# 1. Tus datos (Asegúrate de que sean 5 y 5)
eje_x = ['1', '5000', '10000', '15000', '20000'] # Los ponemos como texto (strings)
eje_y = [78, 60, 56, 44, 42] 

plt.figure(figsize=(10, 6))

# 2. Al ser texto, Python los pondrá a la misma distancia uno de otro
plt.bar(eje_x, eje_y, color='#DCD0FF', edgecolor='#B19CD9')

# 3. Ajustamos el límite del eje Y como pediste
plt.ylim(0, 80)

plt.title('Relació preu-producció', fontsize=14)
plt.xlabel('Rang de producció', fontsize=12)
plt.ylabel('Preu €', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos de las etapas
etapas = ['Bateria (12V)', 'Reg.buck 5V', 'Reg. 3.3V', 'Càrrega']
voltajes = [12, 5, 3.3, 3.3]

# Configuración de colores pastel
color_linea = '#D8BFD8'  # Thistle (Lila pastel)
color_puntos = '#FFB6C1' # Light Pink (Rosa pastel)

ax = plt.axes()
ax.set_facecolor("#FAFAD2") # Un toque crema para contraste

# Dibujar la gráfica de pasos
plt.step(range(len(etapas)), voltajes, where='post', 
         color=color_linea, linewidth=4, 
         marker='o', markersize=10, markerfacecolor=color_puntos, markeredgecolor=color_linea)

# Títulos y etiquetas
plt.title('Salt de voltatge del regulador de tensió', fontsize=14, fontweight='bold', color='#8B4789')
plt.ylabel('Voltaje (V)', fontsize=12, fontweight='bold')
plt.xticks(range(len(etapas)), etapas, fontsize=10)
plt.ylim(0, 14)
plt.grid(True, linestyle=':', alpha=0.6, color='#D8BFD8')

# Etiquetas de valores
for i, v in enumerate(voltajes):
    plt.text(i, v + 0.5, f'{v}V', ha='center', fontweight='bold', color='#8B4789')

plt.tight_layout()
plt.show()
```
